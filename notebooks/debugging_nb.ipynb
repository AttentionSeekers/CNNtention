```python
def load_data(name='CIFAR-10'):
    if name == 'CIFAR-10':
        train_set = torchvision.datasets.CIFAR10(root='../data',
                                                 train=True,
                                                 download=True)
        test_set = torchvision.datasets.CIFAR10(root='../data',
                                                train=False,
                                                download=True
                                                )

    return train_set, test_set
```
